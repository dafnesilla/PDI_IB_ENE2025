```python
import cv2
import matplotlib.pyplot as plt
```

Se llaman a las librerías de cv2 y de matplotlib que son las que necesitaremos en el documento.

```python
img_bgr = cv2.imread('Im004_ACRIMA.jpg')
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

img_hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)

h, s, v = cv2.split(img_hsv)

v_eq = cv2.equalizeHist(v)

img_hsv_eq = cv2.merge([h, s, v_eq])

img_eq_bgr = cv2.cvtColor(img_hsv_eq, cv2.COLOR_HSV2BGR)
img_eq_rgb = cv2.cvtColor(img_eq_bgr, cv2.COLOR_BGR2RGB)
```

En esta celda cargamos la imagen, luego la convertimos a RGB para una mejor visualización convertimos la imagen al espacio de color HSV (hue, saturation, value) donde se hace mas facil manipular el brillo sin afectar su color, se separan estos mismos canales, se mejora el contraste del canal del brillo usando la ecualización del histograma. Se vuelven a unir los canales hsv y se le agrega la ecualización para una imagen mejorada y después se convierte de nuevo a una imagen RGB mejorada con mayor contraste visual.

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title('Imagen original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img_eq_rgb)
plt.title('Imagen mejorada (Ecualización V)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
