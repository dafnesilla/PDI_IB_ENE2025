```python
import cv2
import numpy as np
import matplotlib.pyplot as pl

imagen1 = cv2.imread("/content/chest_PA.jpg", cv2.IMREAD_GRAYSCALE)
imagen2 = cv2.imread("/content/Ultrasonido.jpg", cv2.IMREAD_GRAYSCALE)

imagen2_res = cv2.resize(imagen2, (imagen1.shape[1], imagen1.shape[0]))
alpha = 0.2
beta = 0.8
gamma = 10

imagen_resultado = cv2.addWeighted(imagen1, alpha, imagen2_res, beta, gamma)
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(imagen1, cmap="gray")
plt.title("Imagen Original 1")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(imagen2, cmap="gray")
plt.title("Imagen Original 2")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(imagen_resultado, cmap="gray")
plt.title("Suma Ponderada")
plt.axis("off")

plt.show()
```

```python
imagenes = ["/content/chest_PA.jpg", "/content/Ultrasonido.jpg"]
imagenes_cargadas = [cv2.imread(img, cv2.IMREAD_GRAYSCALE) for img in imagenes]

alto, ancho = imagenes_cargadas[0].shape
imagenes_cargadas = [cv2.resize(img, (ancho, alto)) for img in imagenes_cargadas]

suma = np.zeros((alto, ancho), dtype=np.float32)

for img in imagenes_cargadas:
    suma += img
imagen_promedio = suma / len(imagenes_cargadas)

imagen_promedio = cv2.convertScaleAbs(imagen_promedio)
for i, img in enumerate(imagenes_cargadas):
    plt.subplot(1, len(imagenes), i+1)
    plt.imshow(img, cmap="gray")
    plt.title(f"Imagen {i+1}")
    plt.axis("off")

plt.show()

plt.figure(figsize=(5, 5))
plt.imshow(imagen_promedio, cmap="gray")
plt.title("Suma Promedio")
plt.axis("off")
plt.show()
```

```python
imagen1 = cv2.imread("/content/chest_PA.jpg", cv2.IMREAD_GRAYSCALE)
imagen2 = cv2.imread("/content/Ultrasonido.jpg", cv2.IMREAD_GRAYSCALE)

imagen2 = cv2.resize(imagen2, (imagen1.shape[1], imagen1.shape[0]))

resta = cv2.subtract(imagen2, imagen1)

abs_diff = cv2.absdiff(imagen1, imagen2)

plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.imshow(imagen1, cmap='gray')
plt.title("Imagen 1")
plt.axis("off")

plt.subplot(2, 2, 2)
plt.imshow(imagen2, cmap='gray')
plt.title("Imagen 2")
plt.axis("off")

plt.subplot(2, 2, 3)
plt.imshow(resta, cmap='gray')
plt.title("Resta")
plt.axis("off")

plt.subplot(2, 2, 4)
plt.imshow(abs_diff, cmap='gray')
plt.title("Resta absoluta")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
imagen1 = cv2.imread("/content/chest_PA.jpg", cv2.IMREAD_GRAYSCALE)
imagen2 = cv2.imread("/content/Ultrasonido.jpg", cv2.IMREAD_GRAYSCALE)

imagen2 = cv2.resize(imagen2, (imagen1.shape[1], imagen1.shape[0]))

factor1 = 0.2
factor2 = 3

imagen1_multiplicada = cv2.multiply(imagen1.astype(np.float32), factor1)
imagen1_multiplicada = cv2.convertScaleAbs(imagen1_multiplicada)

imagen2_multiplicada = cv2.multiply(imagen2.astype(np.float32), factor2)
imagen2_multiplicada = cv2.convertScaleAbs(imagen2_multiplicada)

plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.imshow(imagen1, cmap='gray')
plt.title("Imagen 1 Original")
plt.axis("off")

plt.subplot(2, 2, 2)
plt.imshow(imagen1_multiplicada, cmap='gray')
plt.title(f"Imagen 1 x {factor1}")
plt.axis("off")

plt.subplot(2, 2, 3)
plt.imshow(imagen2, cmap='gray')
plt.title("Imagen 2 Original")
plt.axis("off")

plt.subplot(2, 2, 4)
plt.imshow(imagen2_multiplicada, cmap='gray')
plt.title(f"Imagen 2 x {factor2}")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img1 = cv2.imread('/content/chest_PA.jpg', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('/content/Ultrasonido.jpg', cv2.IMREAD_GRAYSCALE)
img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

_, binary1 = cv2.threshold(img1, 0, 255, cv2.THRESH_BINARY)
_, binary2 = cv2.threshold(img2, 0, 255, cv2.THRESH_BINARY)

and_img = cv2.bitwise_and(binary1, binary2)
or_img = cv2.bitwise_or(binary1, binary2)
xor_img = cv2.bitwise_xor(binary1, binary2)
not_img1 = cv2.bitwise_not(binary1)
not_img2 = cv2.bitwise_not(binary2)

titles = ["Imagen 1", "Imagen 2","Binaria 1","Binaria 2", "AND", "OR", "XOR", "NOT (imagen 1)", "NOT (imagen 2)"]
images = [img1, img2, binary1, binary2, and_img, or_img, xor_img, not_img1, not_img2]

plt.figure(figsize=(12, 6))
for i in range(9):
    plt.subplot(3, 3, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')


plt.tight_layout()
plt.show()
```
